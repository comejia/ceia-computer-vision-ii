# Detección de objetos con YOLOv11

YOLO (You Only Look Once) es una familia de modelos de detección de objetos en tiempo real. En esta sección, se muestra cómo utilizar un modelo YOLOv11 para detectar objetos en imágenes.

```python
from ultralytics import YOLO
import cv2

video_path = "./oficina.mp4"
```

```python
# Load a pretrained YOLO11 model
model_n = YOLO("yolo11n.pt")
model_s = YOLO("yolo11s.pt")
model_m = YOLO("yolo11m.pt")
model_l = YOLO("yolo11l.pt")
```

```python
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    raise Exception("Could not open video.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("End of video stream.")
        break

    results = model_s.predict(source=frame, device="cpu", conf=0.7, imgsz=1280, verbose=True)
    annotated_frame = results[0].plot()

    cv2.imshow("YOLO Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

📝 Mientras mas grande el modelo usado, mas tardaará en procesar cada imagen, pero también puede detectar objetos con mayor precisión. Para videos pregrabados, el modelo va a procesar todos los frames, pero para videos en vivo puede exister **perdida de frames** si el modelo tarda mucho en procesar cada frame. Por lo tanto, para videos en vivo es recomendable usar un modelo mas pequeño, como el modelo "s" (small), mientras que para videos pregrabados se puede usar un modelo mas grande, como el modelo "m" (medium).
