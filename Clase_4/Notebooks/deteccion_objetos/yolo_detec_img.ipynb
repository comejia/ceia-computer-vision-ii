```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

imagen_entrada = "parque.png"

model = YOLO("yolo11m.pt")

results = model.predict(source=imagen_entrada, device=0, conf=0.6, imgsz=640, verbose=True)
r = results[0]

annotated = r.plot()[:, :, ::-1]

plt.figure(figsize=(10, 7))
plt.imshow(annotated)
plt.axis("off")
plt.show()

for b in r.boxes:
    cls = int(b.cls[0])
    conf = float(b.conf[0])
    x1, y1, x2, y2 = map(float, b.xyxy[0].tolist())
    print(f"{model.names[cls]} {conf:.3f} [{x1:.1f}, {y1:.1f}, {x2:.1f}, {y2:.1f}]")
```
