```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
import time

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

ruta_entrada = "/content/pose1.jpg"

imagen_base = Image.open(ruta_entrada).convert("RGB")
imagen_base = np.array(imagen_base)

estilo_puntos = mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=20, circle_radius=4)
estilo_lineas = mp_drawing.DrawingSpec(color=(0, 255, 255), thickness=30)

modelos = [0, 1, 2]
resultados_img = []

for mc in modelos:
    imagen_rgb = imagen_base.copy()

    inicio = time.perf_counter()

    with mp_pose.Pose(
        static_image_mode=True,
        model_complexity=mc,
        min_detection_confidence=0.5
    ) as pose:
        resultados = pose.process(imagen_rgb)

    fin = time.perf_counter()
    tiempo_ejecucion = fin - inicio

    if resultados.pose_landmarks:
        mp_drawing.draw_landmarks(
            imagen_rgb,
            resultados.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=estilo_puntos,
            connection_drawing_spec=estilo_lineas
        )

    resultados_img.append((mc, imagen_rgb, tiempo_ejecucion))

    imagen_bgr_salida = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2BGR)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for ax, (mc, img, t) in zip(axes, resultados_img):
    ax.imshow(img)
    ax.set_title(f"model_complexity = {mc}\nTiempo = {t:.4f} s", fontsize=14)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

ruta_entrada = "pose1.jpg"
ruta_salida = "salida_esqueleto.jpg"

imagen_bgr = cv2.imread(ruta_entrada)


imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)
h, w, _ = imagen_rgb.shape

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

estilo_puntos = mp_drawing.DrawingSpec(color=(0, 0, 255), thickness=12, circle_radius=8)
estilo_lineas = mp_drawing.DrawingSpec(color=(0, 255, 255), thickness=18)

with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
    resultados = pose.process(imagen_rgb)

    coords_norm = None
    coords_pix = None

    if resultados.pose_landmarks:
        mp_drawing.draw_landmarks(
            imagen_rgb,
            resultados.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=estilo_puntos,
            connection_drawing_spec=estilo_lineas
        )

        landmarks = resultados.pose_landmarks.landmark

        coords_norm = np.array([
            [lm.x, lm.y, lm.z, lm.visibility]
            for lm in landmarks
        ])

        coords_pix = np.array([
            [lm.x * w, lm.y * h, lm.z, lm.visibility]
            for lm in landmarks
        ])

        for idx, lm in enumerate(landmarks):
            nombre = mp_pose.PoseLandmark(idx).name
            x_px = lm.x * w
            y_px = lm.y * h
            print(f"{idx:02d} {nombre}: x={lm.x:.3f}, y={lm.y:.3f}, z={lm.z:.3f}, vis={lm.visibility:.3f}, x_px={x_px:.1f}, y_px={y_px:.1f}")

plt.figure(figsize=(8, 8))
plt.imshow(imagen_rgb)
plt.axis("off")
plt.show()

imagen_bgr_salida = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2BGR)
cv2.imwrite(ruta_salida, imagen_bgr_salida)
```
