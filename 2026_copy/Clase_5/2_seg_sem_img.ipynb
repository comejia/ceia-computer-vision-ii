# Segmentación **semántica** de objetos con DeepLab

La segmentación semántica es una tarea de visión por computadora que consiste en asignar una etiqueta de clase a cada píxel de una imagen. A diferencia de la segmentación por instancia, donde se segmentan objetos individuales, la segmentación semántica agrupa píxeles que pertenecen a la **misma clase** sin distinguir entre diferentes instancias del mismo objeto.

```python
import torch
import torch.nn as nn
from torchvision.models.segmentation import deeplabv3_resnet50

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
num_classes = 1

model = deeplabv3_resnet50(weights="DEFAULT")
model.classifier[4] = nn.Conv2d(256, num_classes, kernel_size=(1, 1))
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=3e-4, weight_decay=1e-4)

scaler = torch.amp.GradScaler(enabled=(device.type == "cuda"))
```
