# Detección de objetos con YOLOv11

YOLO (You Only Look Once) es una familia de modelos de detección de objetos en tiempo real. En esta sección, se muestra cómo utilizar un modelo YOLOv11 para detectar objetos en imágenes.

```python
from ultralytics import YOLO
import cv2
import time
# import torch

# device = "cuda" if torch.cuda.is_available() else "cpu"
# print(f"Using device: {device}")
device = "cpu"
```

```python
# Load a pretrained custom YOLO11 model
base_path = "../../runs/detect/runs/custom_yo/weights/"
model = YOLO(f"{base_path}/best.pt")

model.to(device)
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video.")
    raise Exception("Could not open video.")

prev_time = 0.0

while True:
    ret, frame = cap.read()

    if not ret:
        print("End of video stream.")
        break

    start_time = time.time()

    results = model.predict(source=frame, device=device, conf=0.7, imgsz=640, verbose=False)

    annotated_frame = results[0].plot()

    end_time = time.time()

    fps = 1.0 / max(end_time - start_time, 1e-6)
    prev_time = fps if prev_time == 0.0 else (prev_time * 0.9 + fps * 0.1)

    cv2.putText(
        annotated_frame,
        f"FPS: {prev_time:.2f}",
        (10, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
    )

    cv2.imshow("YOLO Detection Real Time", annotated_frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord("q") or key == 27:  # 'q' or ESC to quit
        break

cap.release()
cv2.destroyAllWindows()
```
